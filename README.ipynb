This project aims to accurately predict the total travel time of a NYC taxi ride based on features such as pickup datetime, geographic coordinates, and passenger count. Such predictions can support route optimization, traffic analysis, and customer service improvements.



---
 Dataset
Source: Kaggle – NYC Taxi Trip Duration

Key features include:

pickup_datetime

pickup and dropoff coordinates

passenger_count

trip_duration (target)

---
During the cleaning process part of the dataset has been dropped such as:

tolls amount

total amount

rate code and.... due to irrelevancy

---
**Why Random Forest?**

Random Forest was selected as one of the primary models in this project due to its strong performance in structured data and its ability to handle non-linear relationships. Taxi trip durations are influenced by multiple interacting factors such as time of day, distance, and pickup/dropoff locations—relationships that are not necessarily linear or easily separable.
